```python
import math
def cos():
    x=[]
    for i in range(4):
        if i%2==0:
            x.append(int(input('请输入一个整数，表示向量的横坐标，以回车结束')))
        else:
            x.append(int(input('请输入一个整数，表示向量的纵坐标，以回车结束')))
    a=x[0]*x[0]
    c=x[1]*x[1]
    b=a+c
    d=math.sqrt(b)
    e=x[2]*x[2]
    g=x[3]*x[3]
    f=e+g
    h=math.sqrt(f)
    mul=x[0]*x[2]+x[1]*x[3]
    k=d*h
    result=mul/k
    return(result)
z=cos()
print(z)
```
