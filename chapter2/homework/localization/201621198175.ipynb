```python
#task2:
#练习2：仿照实践1，写出由用户指定整数个数，并由用户输入多个整数，并求和的代码。
n = int(input("请输入要求和的整数的个数，回车结束："))
total = 0
for i in range(n):
    num = int(input("请输入一个整数，回车结束："))
    total += num

print(total)

```

```python
#练习4：用户可以输入的任意多个数字，直到输入所有数字的和比当前输入数字小，且输入所有数字的积比当前输入数字的平方大。
l = []
num = int(input("请输入一个不为0的数字，以回车结束："))
total1 = num
total2 = num
l.append(num)
for each in l:
    number = input("请输入一个不为0的数字，以回车结束：")
    total1 += int(number)
    total2 *= int(number)
    l.append(number)
    if (total1 < int(number)) and (total2 > int(number)*int(number)):
        break

print(total1)
print(total2)
    
```

```python
#task4:
#练习 1：仿照求$ \sum_{i=1}^mi + \sum_{i=1}^ni  + \sum_{i=1}^ki$的完整代码，写程序，可求m!+n!+k!
def my_sum():
    total1 = 1
    total2 = 1
    total3 = 1
    m = int(input("请输入一个整数，回车结束："))
    n = int(input("请输入一个整数，回车结束："))
    k = int(input("请输入一个整数，回车结束："))
    for i in range(m):
        total1 *= (i+1)
    #print(total1)
    for j in range(n):
        total2 *= (j+1)
    #print(total2)
    for l in range(k):
        total3 *= (l+1)
    #print(total3)
        
    return(total1+total2+total3)

x = my_sum()

print(x)
        

```

```python
#练习 2：写函数可返回1 - 1/3 + 1/5 - 1/7...的前n项的和。在主程序中，分别令n=1000及100000，打印4倍该函数的和。
def my_sum():
    total = 0
    num = int(input("请输入要求和的数字的个数，回车结束"))
    for n in range(num):
        n += 1
        total += (-1)**(n-1)*(1.0/(2*n-1))
    
    return total
    

print(4*my_sum())
```

```python
#挑战性练习：写程序，可以求从整数m到整数n累加的和，间隔为k，求和部分需用函数实现，主程序中由用户输入m，n，k调用函数验证正确性。
def my_sum():
    total = 0
    m = int(input("请输入起始数字，以回车结束："))
    n = int(input("请输入末尾数字，以回车结束："))
    k = int(input("请输入间隔数字，以回车结束："))
    for i in range(m,(n+1),k):
        total += i
    
    return total

x = my_sum()
print(x)
        
```

```python
#task5:
#练习 1：写函数，求n个随机整数均值的平方根，整数范围在m与k之间（n,m,k由用户输入）。
import random
from math import sqrt
x = []
n = int(input("请输入要求和的数字的个数，回车结束："))
m = int(input("请输入起始数字，回车结束："))
k = int(input("请输入末尾数字，回车结束："))
for i in range(n):
    number = random.randint(m,k+1)
    print(number)
    x.append(number)
    
total = 0
for each in x:
    total += each
ave = total/n*1.0
avr = sqrt(ave)
    
print("随机整数均值的平方根是：",avr)
    
    
```

```python
#练习 2：写函数，共n个随机整数，整数范围在m与k之间，（n,m,k由用户输入）。
#求1：西格玛log(随机整数)，2：西格玛1/log(随机整数)
import random
from math import log
def my_sum():
    x1 = []
    x2 = []
    total1 = 0
    total2 = 0
    n = int(input("请输入要求和的数字的个数，回车结束："))
    m = int(input("请输入起始数字，回车结束："))
    k = int(input("请输入末尾数字，回车结束："))
    
    for i in range(n):
        number = random.randint(m,k+1)
        print(number)
        xon1 = log(number)
        xon2 = -log(number)
        x1.append(xon1)
        x2.append(xon2)
        
    for each1 in x1:
        total1 += each1
        
    for each2 in x2:
        total2 += each2
        
    return(total1,total2)

y1 = my_sum()
y2 = my_sum()

print("西格玛log(随机整数)的值是：",y1[0])
print("西格玛1/log(随机整数)的值是：",y2[1])
        
    
```

```python
#练习 3：写函数，求s=a+aa+aaa+aaaa+aa...a的值，其中a是[1,9]之间的随机整数。例如2+22+222+2222+22222(此时共有5个数相加)，几个数相加由键盘输入。
import random
def my_sum():
    s = 0
    total = 0
    n = int(input("请输入要求和的整数的个数，回车结束："))
    number = random.randint(1,9)
    print("随机整数是：",number)
    for i in range(n):
        s += number*(10**i)
        total += s
    
    return total

x = my_sum()
print("随机整数的和为：",x)

        
        
```
