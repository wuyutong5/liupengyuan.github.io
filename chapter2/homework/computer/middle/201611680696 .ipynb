```python
#写函数，给定符号和行数，如’*’，5，可打印相应行数的如下图形：

def a(b,n):
    for i in range(1,n+1):
        print(' '*(n-i)+(b+' ')*i+' '*(n-i))

b=input("请输入符号:")
n=int(input("请输入行数："))
a(b,n)
```

```python
#打印如下9*9 乘法口诀表，注意每列左侧竖向对齐

for i in range(1,10):
    for j in range(1,10):
        if i>=j:
            print(i,'*',j,'=',i*j,end=' ')
    print()
```

```python
#有一个列表：[1, 2, 3, 4…n]，n=20；请编写代码打印如下规律的输出（不允许直接拷贝并直接打印作为答案，想象n可能是变量）：

numbers=[]
number=[]

for i in range(20):
    numbers.append(i)
    i += 1

for i in range(1,4):
    print('[')
    for j in range(i-1,i+5):
        if j==i-1:
            numbers[j]=(j+'*')
        else:
            print(',',numbers[j])
        print(number)
        j=0
    print(']')

for i in range(4,19):
    print('[')
    for j in range(i-1,i+5):
        if j==i+1:
            print(numbers[j],'*')
        else:
            print(',',numbers[j])
        print(number)
        j=0
    print(']')

for i in range(19,21):
    print('[')
    for j in range(i-1,i+5):
        if j==i+4:
            print(numbers[j],'*')
        else:
            print(',',numbers[j])
        print(number)
        j=0
    print(']')
```
