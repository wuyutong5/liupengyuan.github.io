```python
def ex():
    n=int(input())
    shus1=[]
    m=0
    for j in range(1,n+1):
        for i in range(1,n+1):
            if j%i==0:
                m+=1
            else:
                continue
            if m==2:
                shus1.append(j)
            else:
                continue
    for a in shus1:
        for b in shus1:
            if a+b==n:
                print(a,b)
            else:
                continue

ex()
```
