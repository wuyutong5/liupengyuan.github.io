```python
def lsum(l):
    sum=0
    for i in l:
        sum=sum+i 
    return sum

number=[1,2,3,4,5]
print(lsum(number))

    
    
    
```

```python
def minlist(l):
    min=l[0]
    for n in l:
        if n<min:
            min=n
    return min

number=[1,2,3,4,6,8,6,0]
print(minlist(number))
```

```python
#写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1.
def wherelist(l,n):
    po = 0
    flag=1
    for i in l:
        po=po+1
        if n==i:
            flag=0
            break
    if(flag):
        po=-1
    return po 

number=[1,2,3,4,5,6]
print(wherelist(number,5))
print(wherelist(number,7))
    
    
            
```

```python
#写函数，可求两个向量的夹角余弦值，向量可放在list中。主程序调用该函数。
import math
def cos(n1,n2):
    nsum=0
    mn1=0
    mn2=0
    for i,j in n1,n2:
        mn2=mn2+j*j
        mn1=mn1+i*i
        nsum=nsum+i*j
        #print(i,' ',j,'\n')
    
    #print(nsum) 
    return nsum/(math.sqrt(mn2)*math.sqrt(mn1))

n1a=float(input('第一个向量的x'))
n2a=float(input('第一个向量的y'))
n1b=float(input('第二个向量的x'))
n2b=float(input('第二个向量的y'))
n1=[n1a,n1b]
n2=[n2a,n2b]
print("余弦值为:",cos(n1,n2))

    
```
