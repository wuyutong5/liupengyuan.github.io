```python
numbers=[1,3,5,7,9]

def mysum():
    sum=0
    for n in numbers:
        sum+=n
    return sum
print(mysum())

def mymin():
    min=numbers[0]
    for n in numbers:
        if n<min:
            min=n
    return min
print(mymin())

def mylocate(find):
    if find in numbers:
        i=0
        for n in numbers:
            if find==n:
                return i
            else:
                i=i+1
    else:
        return -1
print(mylocate(1))
print(mylocate(2))
```

```python
import math
a=[0,2]
b=[3,4]
def cosjiajiao():
    fenzi=a[0]*b[0]+a[1]*b[1]
    fenmu=math.sqrt(a[0]*a[0]+a[1]*a[1])*math.sqrt(b[0]*b[0]+b[1]*b[1])
    result=fenzi/fenmu
    return result
print(cosjiajiao())
```
