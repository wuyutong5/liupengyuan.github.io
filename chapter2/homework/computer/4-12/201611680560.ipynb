写函数，返回一个list中所有数字的和

```python
def he(numbers):
    n=len(numbers)
    i=0
    sum=0
    for i in range(n):
        sum += numbers[i]
        i += 1
    return sum

aaa=[1,2,3,4,5,6]
print(he(aaa))
```

写函数，返回一个list中的最小值

```python
def small(numbers):
    n=len(numbers)
    i=0
    x=numbers[0]
    for i in range(n):
        if x > numbers[i]:
            x=numbers[i]
    return x

hhh=[2,5,7,3,5]
print(small(hhh))
```

写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1.

```python
def position(ls,x):
    n = len(ls)
    p = 0
    for i in range(n):
        if ls[i] == x:
            p = i + 1
            break
    if p == 0:
        return -1
    else:
        return p
    
ls = [1,2,3,4,5,6]
print(ls)
menber = int(input("input the menber you want to know the position:"))
print(position(ls,menber))
```

写函数，可求两个向量的夹角余弦值，向量可放在list中。主程序调用该函数。

```python
def cos(x,y):
    if(len(x)!=len(y)):
        print('error input,x and y is not in the same space')
        return;
    r1 = 0.0
    r2 = 0.0
    r3 = 0.0
    for i in range(len(x)):
        r1 += x[i] * y[i]   
        r2 += x[i] ** 2   
        r3 += y[i] ** 2 
        r = r1 / ((r2 * r3)**0.5)
    print("result is ",r) 

cos([2,1],[1,1])
```
