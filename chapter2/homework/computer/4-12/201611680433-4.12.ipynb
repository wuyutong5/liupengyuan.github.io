练习2

```python
def funxsum(numbers):
    sum=0
    for i in numbers:
        sum+=i
    return sum

numbers=[]

n=int(input('请输入求和的数字个数：'))

for i in range(n):
    m=int(input('请输入数字：'))
    numbers.append(m)

print('所有数字的和：',funxsum(numbers))
```

练习3

```python
def funxsmall(numbers,n):
    min=numbers[0]
    for i in range(n):
        if min>numbers[i]:
            min=numbers[i]
    return min

numbers=[]

n=int(input('请输入数字的个数：'))

for i in range(n):
    m=int(input('请输入数字：'))
    numbers.append(m)
    
print('序列中最小的数字为：',funxsmall(numbers,n))
```

练习4

```python
def weizhi(numbers,n,m):
    for i in range(n):
        if m==numbers[i]:
            return i+1;
        else:
            continue;
    if i==n-1:
        return -1;

numbers=[]

n=int(input('请输入数字的个数：'))

for i in range(n):
    m=int(input('请输入数字：'))
    numbers.append(m)
    
x=int(input('求位置的数字：'))
    
print('该数字在序列中的位置为：',weizhi(numbers,n,x))
        
```

练习5

```python
def yuxian(numbers1,numbers2):
    
    ji_s=0
    sum1_s=0
    sum1_a=0
    sum2_s=0
    sum2_a=0
    
    for i in range(2):
        ji_t=numbers1[i]*numbers2[i]
        ji_s+=ji_t
        
    for i in range(2):
        sum1_t=numbers1[i]**2
        sum1_s+=sum1_t
    sum1_a=sum1_s**(1/2)
    
    for i in range(2):
        sum2_t=numbers2[i]**2
        sum2_s+=sum2_t
    sum2_a=sum2_s**(1/2)
    
    zhi=ji_s/(sum1_a*sum2_a)
    return zhi

numbers1=[]
numbers2=[]

for i in range(2):
    m=int(input('请输入第一个向量：'))
    numbers1.append(m)
    
for i in range(2):
    m=int(input('请输入第二个向量：'))
    numbers2.append(m)
    
print('余弦值是：',yuxian(numbers1,numbers2))
```
