将前面几章用while循环的习题，用for循环实现，并尽量写成函数。

```python
def multiple(end):
    i=1
    mul=1
    for i in range(1,end+1) :
        mul*=i
        i+=1
    return mul

n1=int(input("please enter a numbers"))
n2=int(input("please enter a numbers"))
n3=int(input("please enter a numbers"))
print('the result is: ',multiple(n1)+multiple(n2)+multiple(n3))
```

```python
def add(end):
    bla=0
    for temp in range (1,2*end,2) :
        bla += 1/temp
    return bla

print(4*add(1000),4*add(100000),sep='    ')
```

```python
def operation(a,b,oper) :
    if oper=='add' :
        total=0
        for i in range(a,b+1):
            total+=i
        print(total)
    elif oper=="power" :
        multiple=1
        for i in range(a,b+1) :
                multiple*=i
        print(multiple)
    elif oper=="reminder" :
        print(b%a)
    else :
        print(b//a)
    
m=int(input("please enter m"))
n=int(input("please enter n"))
oper=input("please enter the operation you want(add, power devision or reminder)")
if m<=n :
    a=m
    b=n
else :
    a=n
    b=m
operation(a,b,oper)
```

写函数，返回一个list中所有数字的和

```python
def add(_list):
    total=0
    for i in _list:
        total+=i
    return total

numbers=[number for number in range(1,5)]
print(numbers)
add(numbers)
```

写函数，返回一个list中的最小值

```python
def smaller(a,b):
    if a > b :
        return b
    else:
        return a
    
def min_number(numbers):
    p=numbers[0]
    for i in range(1,len(numbers)):
        p=smaller(numbers[i],p)
    return p    
    
    
_list=[1,44,-34,77,9,0,-102,76]
print(min_number(_list))
```

写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1

```python
def findset(word,numbers):
    if word not in numbers:
        return -1
    else:
        k=0
        for i in numbers:
            k+=1
            if word==i:
                break;
        print("the position of ",word,"is ",k)
        
words = []
n = int(input('please enter number of the words : '))
for i in range(n):
    word = input('please enter a word')
    words.append(word)
word=input("please enter the word you want to find : ")
findset(word,words)
```
