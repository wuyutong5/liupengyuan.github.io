```python
import math,random
def pingfan():
    n=int(input('请输入你要求的次数'))
    k=int(input('请输入你的上限'))
    j=int(input('请输入你的下限'))
    total=0
    for i in range(n):
        total=total+random.randint(k,j)
    m=total/n
    return math.sqrt(m)
print(pingfan())
```

```python
numbers=[67,45,67,4,37,65]
def mix_1():
    t=len(numbers)
    mix=numbers[0]
    for i in range(1,t):
        if numbers[i] < mix:
            mix=numbers[i]
    return mix
print(mix_1())
```

```python
numbers=[45,56,48,34,67,56]
def xunzhao():
    n=int(input('请输入你要寻找的元素'))
    t=len(numbers)
    i=0
    while i < t:
        if n==numbers[i]:
            return i+1
        elif i==t-1:
            return -1
        i=i+1
print(xunzhao())
            
        
```
