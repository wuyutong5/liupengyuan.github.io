```python
#1、随机生成1万个整数，范围在0-10万之间，分别进行简单选择排序、快速排序（自行递归实现的）以及内置sort函数3种排序，打印出3种排序的运行时间。

import time
import random

def simple_sort(seq):
    start_time = time.time()
    for i in range(len(seq)-1):
        min = i
        for j in range(i+1,len(seq)):
            if seq[j] < seq[min]:
                min = j
        seq[i],seq[min] = seq[min],seq[i]
    
    end_time = time.time()
    print('The total time of the simple sort is: ', end_time-start_time)
    
def recursive_sort(seq):
    left_seq = []
    right_seq = []
    k = seq[0]
    start_time = time.time()
    for number in seq[1:]:
        if number <= k:
            left_seq.append(number)
        else:
            right_seq.append(number)
    if left_seq:
        left_seq = recursive_sort(left_seq)
    if right_seq:
        right_seq = recursive_sort(right_seq)
        
    end_time = time.time()
    
    print('The total time of the recursive sort is : ', end_time-start_time)
    return left_seq + [k] + right_seq
    
def inside_sort(seq):
    start_time = time.time()
    seq.sort()
    end_time = time.time()
    print('The total time of the inside sort is : ', end_time-start_time)
    
seq = [random.randint(0,100001) for _ in range(10000)]
simple_sort(seq)
recursive_sort(seq)
inside_sort(seq)
    
    
```

```python
#2、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

import random

def sort(seq):
    int_freq_pairs = []
    seq.sort()
    for integer in seq:
        for item in int_freq_pairs:
            if integer == item[0]:
                item[1]+=1
                break
        else:
            int_freq_pairs.append([integer,1])
        
    return int_freq_pairs

seq = [random.randint(0,100001) for _ in range(10000)]
print(sort(seq))
    
```
