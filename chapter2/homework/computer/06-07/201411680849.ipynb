```python
#1、随机生成1万个整数，范围在0-10万之间，分别进行简单选择排序、快速排序
#（自行递归实现的）以及内置sort函数3种排序，打印出3种排序的运行时间。
import random,time,bisect

def random_numbers(n):
    numbers=[]
    num=0
    for i in range(n):
        a=random.randint(0,100000)
        numbers.append(a)
    return numbers

def sort_simple_selection(seq):
    intnum=0
    total_int_number=len(seq)
    print('整数的总数为：',total_int_number)
    start_time=time.time()
    for i in range(len(seq)-1):
        min = i
        for j in range(i+1, len(seq)):
            if seq[j] < seq[min]:
                min = j
        b=seq[min]
        seq[min]=seq[i]
        seq[i]=b
        if intnum % 1000 == 0:
            end_time=time.time()
            print('...当前已经处理到第{}个...已经处理了{:.2f}秒...'.format(intnum, end_time - start_time))        
        intnum += 1

def quick_sort(seq):
    left_seq=[]
    right_seq=[]
    p=seq[0]
    for i in range(1,len(seq)):
        if seq[i]<=p:
            left_seq.append(seq[i])
        else:
            right_seq.append(seq[i])
    if left_seq:
        left_seq=quick_sort(left_seq)
    if right_seq:
        right_seq=quick_sort(right_seq)
    return  left_seq + [p] + right_seq

def main(n):
    numbers=random_numbers(10000)
    if(n==1):
        sort_simple_selection(numbers)
        print('简单选择排序：',numbers)
    elif(n==2):
        start_time=time.time()
        sorted_numbers = quick_sort(numbers)
        end_time = time.time()
        print('快速排序：\n用时：',end_time-start_time,'秒\n排序结果为：\n',sorted_numbers)
    else:
        start_time = time.time()
        numbers.sort()
        end_time = time.time()
        print('内置sort排序：\n用时：',end_time-start_time,'秒\n排序结果为：\n', numbers)

main(1)
main(2)
main(3)
```

```python
#2、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import random

def random_numbers(n):
    numbers=[]
    num=0
    for i in range(n):
        a=random.randint(0,100000)
        numbers.append(a)
    return numbers

def count_nums_freq(numbers):
    nums_freq_pairs=[]
    for number in numbers:
        for item in nums_freq_pairs:#一组item，前一个为词，后一个为词频
            if number==item[0]:
                item[1]+=1
                break
        else:
            nums_freq_pairs.append([number,1])
    nums_freq_pairs.sort()
    return nums_freq_pairs

def main():
    numbers=random_numbers(10000)
    pairs=count_nums_freq(numbers)
    print(pairs)
    
    
main()
```
