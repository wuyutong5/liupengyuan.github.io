1、随机生成1万个整数，范围在0-10万之间，分别进行简单选择排序、快速排序（自行递归实现的）以及内置sort函数3种排序，打印出3种排序的运行时间。

```python
def sort_simple_selection(seq):
    for i in range(len(seq)-1):
        min = i
        for j in range(i+1, len(seq)):
            if seq[j] < seq[min]:
                min = j
        seq[i], seq[min] = seq[min], seq[i]

def quick_sort(seq):
    left_seq =  []
    right_seq = []
    p=seq[0]
    for number in seq[1:]:
        if number <= p:
            left_seq.append(number)
        else:
            right_seq.append(number)
    if left_seq:
        left_seq = quick_sort(left_seq)
    if right_seq:
        right_seq = quick_sort(right_seq)

    return  left_seq + [p] + right_seq


import random
import time

numbers = random.sample(range(100000),10000)

start1=time.time()
n=quick_sort(numbers)
time1=time.time()
print('快速排序使用时间',start1-time1)

start2=time.time()
m=sorted(numbers)
time2=time.time()
print('内置sort使用时间',start2-time2)

start3=time.time()
sort_simple_selection(numbers)
time3=time.time()
print('简单选择排序使用时间',start3-time3)
```

2、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

```python
import random

numbers = [random.randint(0,100001) for _ in range(10000)]

m=sorted(numbers)

num_freq_pairs = []
for num in m:
    number = m.count(num)
    num_freq_pairs.append([num, number])
print(num_freq_pairs)
```
