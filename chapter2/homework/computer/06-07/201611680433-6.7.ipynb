随机生成1万个整数，范围在0-10万之间，分别进行简单选择排序、快速排序（自行递归实现的）以及内置sort函数3种排序，打印出3种排序的运行时间。

```python
import random
import time

def sort_simple_selection(seq):
    start_time = time.time()
    for i in range(len(seq)-1):
        min = i
        for j in range(i+1, len(seq)):
            if seq[j] < seq[min]:
                min = j
        seq[i], seq[min] = seq[min], seq[i]
    end_time = time.time()
    print('处理了{:.2f}秒'.format(end_time - start_time))



numbers=[random.randint(0,100001) for _ in range(10000)]
sort_simple_selection(numbers)
print(numbers)
```

```python
import random
import time

def quick_sort(seq):
    start_time = time.time()
    left_seq =  []
    right_seq = []
    p=seq[0]
    for number in seq[1:]:
        if number <= p:
            left_seq.append(number)
        else:
            right_seq.append(number)
    if left_seq:
        left_seq = quick_sort(left_seq)
    if right_seq:
        right_seq = quick_sort(right_seq)
    end_time = time.time()
    print('处理了{:.2f}秒'.format(end_time - start_time))
    return  left_seq + [p] + right_seq

numbers=[random.randint(0,100001) for _ in range(10000)]
sorted_numbers = quick_sort(numbers)
print(sorted_numbers)
```

```python
numbers=[random.randint(0,100001) for _ in range(10000)]
nums = sorted(numbers)
print('nums = ', nums)
print('numbers in sorted(numbers) = ', numbers)
numbers.sort()
print('numbers after numbers.sort() = ', numbers)
```
