```python
#练习2
import random,time

nums = [] 
num_freq_pairs = []
start_time = time.time()
for i in range(10000):
    nums.append(random.randrange(100000))
nums = sorted(nums)
for num in nums:
        for item in num_freq_pairs:
            if num == item[0]:
                item[1] += 1
                break
        else:
            num_freq_pairs.append([num, 1])
            
print(num_freq_pairs)
end_time = time.time()
print(end_time - start_time)
```
