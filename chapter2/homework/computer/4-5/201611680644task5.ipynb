```python
#求n个随机整数均值的平方根，整数范围在m与k之间

m=int(input('请输入一个整数'))
k=int(input('请输入一个整数'))
n=int(input('请输入一个整数'))
if m>k:
    m,k=k,m
import random
add=0
i=0
while i<n:
    aver=random.randrange(m,k)
    add=aver+add
    i+=1
aver=add/n
import math
x=math.sqrt(aver)
print(x)
```

```python
#共n个随机整数，整数范围在m与k之间，求西格玛log（随机整数）及西格玛1/log（随机整数）
m=int(input('请输入一个整数'))
k=int(input('请输入一个整数'))
n=int(input('请输入一个整数'))
if m>k:
    m,k=k,m
import random
import math
add1=0
add2=0
i =0
while i<n:
    number=random.randrange(m,k)
    while number==1:
        number=random.randrange(m,k)
    x=math.log10(number)
    y=1/x
    add1=add1+x
    add2=add2+y
    i=i+1  
print(add1)
print(add2)
```

```python
#求s=a+aa+aaa+aaaa+aa...a的值，其中a是[1,9]之间的随机整数，几个数由键盘键入

n=int(input('请输入一个整数'))
import random
a=random.randrange(1,9)
i=0
b=a
s=a
while i<n-1:
    x=b
    b=x*10+a
    s=s+b
    i=i+1
print(s)
```
